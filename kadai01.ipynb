```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from mylibdir.class_ica import ICA

#データの読み込み
path='report_kadai1_data1/dat1.txt'
with open(path) as f:
    l=f.readlines()
    data1=[]
    for i in range(len(l)):
        l[i]=l[i].strip()
        data1.append(float(l[i]))
        
path='report_kadai1_data1/dat2.txt'
with open(path) as f:
    l=f.readlines()
    data2=[]
    for i in range(len(l)):
        l[i]=l[i].strip()
        data2.append(float(l[i]))
data1=np.array(data1)
data2=np.array(data2)
```

```python
plt.plot(data1)
plt.title("dat1")
plt.show()
```

```python
plt.plot(data2)
plt.title("dat2")
plt.show()
```

```python
data=[data1, data2]
y=ICA(data).run() # ICAの実行
y=[y_i for y_i in np.asarray(y)]
```

```python
#結果
plt.plot(y[0])
plt.title("signal1")
plt.show()
```
